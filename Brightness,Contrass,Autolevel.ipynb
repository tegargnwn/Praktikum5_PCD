Brightness, Contrass, Autolevel Contrass

Import dependency

```python
#import Library
import numpy as np
import imageio.v2
import matplotlib.pyplot as plt
```

Membaca gambar

```python

img = imageio.v2.imread("fototegar.jpeg") #Membaca file gambar dan mengubahnya menjadi representasi numerik
```

Mendapatkan resolusi dan type dari gambar

```python
#Menambahkan resolusi tinggi dan lebar sebuah gambar dan type gambar
img_height = img.shape[0]
img_width = img.shape[1]
img_channel = img.shape[2]
img_type = img.dtype
```

Brightness Grayscale

Membuat variabel img_brightness untuk menampung hasil

```python
#membuat shape kosong
img_brightness = np.zeros(img.shape, dtype=np.uint8)
```

Melakukan penambahan brightness dengan nilai yg menjadi parameter

```python
def brighter(nilai): #membuat fungsi baru dengan parameter nilai
    for y in range(0, img_height): #y untuk tinggi
        for x in range(0, img_width): #x untuk lebar
            red = img[y][x][0] #menambahkan variable channel red
            green = img[y][x][1] #menambahkan variable channel green
            blue = img[y][x][2] #menambahkan variable channel blue
            gray = (int(red) + int(green) + int(blue)) / 3 #rumus channel grey
            gray += nilai #menambahkan nilai ke dalam variabel gray
            if gray > 255: #menentukan batas atas channel gray
                gray = 255
            if gray < 0: #menentukan batas bawah channel gray
                gray = 0
            img_brightness[y][x] = (gray, gray, gray) #mengubah warna dari sebuah pixe gambar RGB menjadi gray
```

Menampilkan beberapa hasil dengan nilai brightness -100 dan 100

```python
brighter(-100) #menambahkan nilai Brightness
plt.imshow(img_brightness) #menampilkan gambar yang kecerahannya sudah diatur sebelumnya
plt.title("Brightness -100") #menambahkan judul pada channel gambar
plt.show() #library matplotlib untuk menampilkan gambar

brighter(100) #menambahkan nilai Brightness
plt.imshow(img_brightness) #menampilkan gambar yang kecerahannya sudah diatur sebelumnya
plt.title("Brightness 100") #menambahkan judul pada channel gambar
plt.show() #library matplotlib untuk menampilkan gambar
```

Brightness RGB

Membuat variabel img_rgbbrightness untuk menampung hasil

```python
#membuat shape kosong
img_rgbbrightness = np.zeros(img.shape, dtype=np.uint8)
```

Melakukan penambahan brightness dengan nilai yg menjadi parameter

```python
def rgbbrighter(nilai): #membuat fungsi baru dengan parameter nilai
    for y in range(0, img_height): #y untuk tinggi
        for x in range(0, img_width): #x untuk lebar
            red = img[y][x][0] #menambahkan variable channel red
            red += nilai #menambahkan nilai ke dalam variabel red
            if red > 255: #membuat batas atas nilai pixel channel red
                red = 255
            if red < 0: #membuat bawah atas nilai pixel channel red
                red = 0
            green = img[y][x][1]  #menambahkan variable channel green
            green += nilai #menambahkan nilai ke dalam variabel green
            if green > 255: #membuat batas atas nilai pixel channel green
                green = 255
            if green < 0: #membuat bawah atas nilai pixel channel green
                green = 0
            blue = img[y][x][2] #menambahkan variable channel blue
            blue += nilai #menambahkan nilai ke dalam variabel blue
            if blue > 255:  #membuat batas atas nilai pixel channel blue
                blue = 255
            if blue < 0:  #membuat bawah atas nilai pixel channel blue
                blue = 0
            img_rgbbrightness[y][x] = (red, green, blue) #menambahkan kecerahan pada pixel gambar dengan warna RGB
```

Menampilkan beberapa hasil dengan nilai brightness -100 dan 100

```python
rgbbrighter(-100) #menambahkan nilai Brightness
plt.imshow(img_rgbbrightness) #menampilkan gambar yang kecerahannya sudah diatur sebelumnya
plt.title("Brightness -100") #menambahkan judul pada channel gambar
plt.show() #library matplotlib untuk menampilkan gambar

rgbbrighter(100) #menambahkan nilai Brightness
plt.imshow(img_rgbbrightness) #menampilkan gambar yang kecerahannya sudah diatur sebelumnya
plt.title("Brightness 100") #menambahkan judul pada channel gambar
plt.show() #library matplotlib untuk menampilkan gambar
```

Contrass

1. Membuat variabel img_contrass untuk menampung hasil

```python
#membuat shape kosong
img_contrass = np.zeros(img.shape, dtype=np.uint8)
```

2. Melakukan penambahan contrass dengan nilai yg menjadi parameter

```python
def contrass(nilai): #membuat fungsi baru dengan parameter nilai
    for y in range(0, img_height): #y sebagai tinggi
        for x in range(0, img_width): #x sebagai lebar
            red = img[y][x][0] #menambahkan variable channel red
            green = img[y][x][1] #menambahkan variable channel green
            blue = img[y][x][2] #menambahkan variable channel blue
            gray = (int(red) + int(green) + int(blue)) / 3 #rumus channel grey
            gray *= nilai #mengalikan setiap nilai piksel dalam gambar dengan "nilai" yang diberikan
            if gray > 255: #membuat batas atas channel
                gray = 255
            img_contrass[y][x] = (gray, gray, gray) #menambahkan kecerahan pada pixel gambar dengan warna grey
```

Menampilkan beberapa hasil dengan nilai contrass 50 dan 100

```python
contrass(2) #menambahkan nilai contrass
plt.imshow(img_contrass) #menampilkan gambar yang kecerahannya sudah diatur sebelumnya
plt.title("Contrass 2") #menambahkan judul pada channel
plt.show() #menampilkan gambar

contrass(3)  #menambahkan nilai contrass
plt.imshow(img_contrass) #menampilkan gambar yang kecerahannya sudah diatur sebelumnya
plt.title("Contrass 3") #menambahkan judul pada channel
plt.show() #menampilkan gambar
```

Contrass Autolevel

1. Membuat variabel img_contrass untuk menampung hasil

```python
#membuat shape kosong
img_autocontrass = np.zeros(img.shape, dtype=np.uint8)
```

2. Melakukan penambahan contrass secara otomatis

```python
def autocontrass():  #membuat fungsi baru dengan parameter nilai
    xmax = 300 #nilai dari varible
    xmin = 0
    d = 0
    # Mendapatkan nilai d, dimana nilai d ini akan berpengaruh pada hitungan
    # untuk mendapatkan tingkat kontras
    for y in range(0, img_height): #y untuk tinggi
        for x in range(0, img_width): #x untuk lebar
            red = img[y][x][0] #membuat variable channel red
            green = img[y][x][1] #membuat variable channel green
            blue = img[y][x][2] #membuat variable channel blue
            gray = (int(red) + int(green) + int(blue)) / 3 #rumus variable channel grey
            if gray < xmax: #batas max nilai gray
                xmax = gray
            if gray > xmin: #batas min nilai gray
                xmin = gray
    d = xmin-xmax #rumus nilai d
    for y in range(0, img_height): #y untuk tinggi
        for x in range(0, img_width): #x untuk lebar
            red = img[y][x][0] #membuat variable channel red
            green = img[y][x][1] #membuat variable channel green
            blue = img[y][x][2] #membuat variable channel blue
            gray = (int(red) + int(green) + int(blue)) / 3 #rumus variable channel grey
            gray = int(float(255/d) * (gray-xmax)) #rumus dari contrass channel grey
            img_autocontrass[y][x] = (gray, gray, gray) #menambahkan kecerahan pada pixel gambar dengan warna grey
```

3. Menampilkan hasil autolevel contrass

```python
autocontrass() #perintah auto contrass
plt.imshow(img_autocontrass) #menampilkan gambar yang kecerahannya sudah diatur sebelumnya
plt.title("Contrass Autolevel") #menambahkan judul
plt.show() #menampilkan gambar
```
