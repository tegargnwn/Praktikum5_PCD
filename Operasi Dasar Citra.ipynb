```python
#import library
import matplotlib.pyplot as plt
%matplotlib inline

from skimage import data
from skimage.io import imread
from skimage.color import rgb2gray 
from skimage.util import invert
import cv2

import numpy as np
```

Percobaan 1 - Cropping Image

```python
astronautImage = data.astronaut() # Mengambil gambar "rocket" dari dataset bawaan "skimage" dan menyimpannya ke dalam variabel "astronautImage"
cameraImage = data.camera() # Mengambil gambar "camera" dari dataset bawaan "skimage" dan menyimpannya ke dalam variabel "cameraImage"

astroCropped = astronautImage.copy() #menduplikasi gambar
astroCropped = astroCropped[0:256,64:320] #mengambil bagian gambar lalu menyimpannya

cameraCropped = cameraImage.copy() #menduplikasi gambar
cameraCropped = cameraCropped[64:256,128:320] #mengambil bagian gambar lalu menyimpannya

print('Astro Ori Shape : ',astronautImage.shape) # Mencetak dimensi matriks gambar
print('Astro Crop Shape : ',astroCropped.shape) # Mencetak dimensi matriks gambar

print('Camera Ori Shape : ',cameraImage.shape) # Mencetak dimensi matriks gambar
print('Camera Crop Shape : ',cameraCropped.shape) # Mencetak dimensi matriks gambar

fig, axes = plt.subplots(2, 2, figsize=(12, 12)) # Membuat plot 2x2 pada ukuran 12x12 inci
ax = axes.ravel() # Meratakan variabel "axes" menjadi array satu dimensi dan menyimpannya ke dalam variabel "ax"

ax[0].imshow(astronautImage) # Menampilkan gambar
ax[0].set_title("Citra Input 1") # Memberikan judul

ax[1].imshow(cameraImage, cmap='gray') # Menampilkan gambar
ax[1].set_title('Citra Input 2') # Memberikan judul

ax[2].imshow(astroCropped) # Menampilkan gambar
ax[2].set_title("Citra Output 1") # Memberikan judul

ax[3].imshow(cameraCropped, cmap='gray') # Menampilkan gambar
ax[3].set_title('Citra Output 2') # Memberikan judul
```

```python
# Load gambar dari local drive
loadImage = imread(fname="fototegar2.jpeg")
loadImagecropped = loadImage.copy() # Menduplikasi gambar
loadImagecropped = loadImagecropped[200:900,500:1500] #mengambil bagian gambar lalu menyimpannya

fig, axes = plt.subplots(1, 2, figsize = (10, 10)) # Membuat plot 1x2 pada ukuran 10x10 inci
ax = axes.ravel()  # Meratakan variabel "axes" menjadi array satu dimensi dan menyimpannya ke dalam variabel "ax"

ax[0].imshow(loadImage) #menampilkan gambar
ax[0].set_title("Citra Input 1") # memberi judul pada gambar
ax[1].imshow(loadImagecropped) #menampilkan gambar
ax[1].set_title("Citra Output 2") # memberi judul pada gambar

fig.tight_layout
plt.show()
```

```python
#Fungsi invert digunakan untuk membuat gambar negatif dari citra input
inv = invert(loadImagecropped)
#mencetak bentuk (shape) dari citra input dan citra output
print('Shape Input : ', loadImagecropped.shape)
print('Shape Output : ',inv.shape)
# Membuat figure subplot
fig, axes = plt.subplots(2, 2, figsize=(12, 12))
ax = axes.ravel()

ax[0].imshow(loadImagecropped)
ax[0].set_title("Citra Input")

ax[1].hist(loadImagecropped.ravel(), bins=256)
ax[1].set_title('Histogram Input')

ax[2].imshow(inv)
ax[2].set_title('Citra Output (Inverted Image)')

ax[3].hist(inv.ravel(), bins=256)
ax[3].set_title('Histogram Output')
```

```python
copyimage = loadImagecropped = loadImage.copy().astype(float)
brightness = 100
cerah_bright = np.clip(copyimage + brightness, 0, 255).astype(np.uint8)

# Membuat figure subplot  
fig, axes = plt.subplots(2, 2, figsize=(8, 8))
ax = axes.ravel()
ax[0].imshow(loadImage, cmap='gray') # menampilkan citra asli hasil crop dari kamera pada subplot pertama
ax[0].set_title("Citra Input") # memberikan judul pada subplot pertama

ax[1].hist(loadImage.ravel(), bins=256) # menampilkan histogram dari citra asli pada subplot kedua
ax[1].set_title('Histogram Input') # memberikan judul pada subplot kedua

ax[2].imshow(cerah_bright, cmap='gray') # menampilkan citra hasil penambahan nilai kecerahan pada subplot ketiga
ax[2].set_title('Citra Output (Brightnes)') # memberikan judul pada subplot ketiga

ax[3].hist(cerah_bright.ravel(), bins=256) # menampilkan histogram dari citra hasil penambahan nilai kecerahan pada subplot keempat
ax[3].set_title('Histogram Output') # memberikan judul pada subplot keempat
```
